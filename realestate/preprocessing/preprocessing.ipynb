```python
import pickle
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import TargetEncoder
```

```python
database = pd.read_csv('../../data/house_prices.csv')
```

```python
database
```

```python
database.isnull().sum()[database.isnull().sum()>0]
```

```python
database['date'] = pd.to_datetime(database['date'], format=r'%Y%m%dT%H%M%S')  # ou infer_datetime_format=True
```

```python
database['year'] = database['date'].dt.year
```

```python
database['month'] = database['date'].dt.month
```

```python
database['sin_month'] = np.sin(2 * np.pi * database['month']/12)
```

```python
database['cos_month'] = np.cos(2 * np.pi * database['month']/12)
```

```python
database = database.drop(columns=['id','date','month'])
```

```python
corr_matrix = database.corr()

plt.figure(figsize=(22,22))

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)

plt.title("Correlation Matrix")

plt.show()
```

```python
x_prices = database.iloc[:,list(range(1,22))].values
```

```python
y_prices = database.iloc[:,0].values
```
